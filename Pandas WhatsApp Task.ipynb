```python
import pandas as pd
```

```python
df=pd.read_csv('DSC WhatsApp Chat.txt',header=None,error_bad_lines=False,encoding='utf8')
df
```

```python
df.drop(0)
df.columns=['Date','Chat']
```

```python
df.head(5)
```

```python
Message= df['Chat'].str.split(' -', n = 1, expand= True)
df['Time']= Message[0]
df['Text']= Message[1]
```

```python
Message1= df['Text'].str.split(':', n = 1, expand= True)

df['Sender']=Message1[0]
df['Message']=Message1[1]
```

```python
df['Message'] = df['Message'].str.replace('<media omitted>','MediaShared')
df['Message'] = df['Message'].str.replace('this message was deleted','DeletedMsg')
data = df.drop(['Chat','Text'],axis=1)
new_data = data.dropna()
new_data
```

```python
new_data.to_csv('Exported_Chat.csv',index= False)
```

### Solution to Questions

Who sent the most messages overall?

```python
import pandas as pd
data= pd.read_csv('Exported_Chat.csv')
data.head(10)
```

```python
data['Sender'].value_counts().head(1)
```

Get the individual that sent most messages per month

```python
# Get the dataframe for each month
august = data[:529]
august
```

```python
september = data[529:]
september
```

```python
# Get the person that sent most messages each month

august['Sender'].value_counts().head(1)
```

```python
september['Sender'].value_counts().head(1)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(context="paper")
```

```python
total_no_of_messages_sent_in_august = pd.DataFrame(august['Sender'].value_counts()).reset_index()
total_no_of_messages_sent_in_august
```

```python
# Total no of people vs total no of messages sent in August
sns.barplot(x='Sender',y='index',data=total_no_of_messages_sent_in_august)
```

```python
total_no_of_messages_sent_in_september = pd.DataFrame(september['Sender'].value_counts()).reset_index()
total_no_of_messages_sent_in_september
```

```python
# Total no of people vs total no of messages sent in September
sns.barplot(x='Sender',y='index',data=total_no_of_messages_sent_in_september)
```

```python
total_number_of_messages_by_each_individual= pd.DataFrame(data['Sender'].value_counts()).reset_index()
total_number_of_messages_by_each_individual
```

```python
# Total no of people vs total no of messages sent by each person
sns.barplot(x='Sender',y='index',data=total_number_of_messages_by_each_individual)
```
